```python
from PIL import Image as PILImage
import os
import pandas as pd
from IPython.display import display

def image_formatter(img_path):
    return f'<img src="{img_path}" width="100">'

def process_images(folder_path):
    # Створення пустого DataFrame для зберігання інформації
    columns = ['Ім’я файлу', 'Кольорова модель', 'Формат', 'Кількість каналів',
               'Розмір у MB', 'Ширина', 'Висота', 'Повний шлях', 'Зменшене зображення']
    image_data = pd.DataFrame(columns=columns)

    # Рекурсивний обхід файлів у папці та її підпапках
    for root, dirs, files in os.walk(folder_path):
        for filename in files:
            if filename.endswith(('.png', '.jpg', '.jpeg', '.gif')):
                file_path = os.path.join(root, filename)
                
                # Відкриття зображення за допомогою Pillow
                img = PILImage.open(file_path)
                
                # Отримання основної інформації
                img_format = img.format
                color_mode = img.mode
                channels = len(img.getbands())
                size_mb = round(os.path.getsize(file_path) / (1024 * 1024), 2)
                width, height = img.size
                full_path = os.path.abspath(file_path)

                # Зміна розмірів зображення
                resized_img = img.resize((100, 100))  # Зміна розмірів на 100x100 (можна вказати інший розмір)

                # Збереження зменшеного зображення в тимчасовий файл
                temp_img_path = f"temp_{filename}"
                resized_img.save(temp_img_path)

                # Додавання інформації до DataFrame
                image_data = image_data.append({
                    'Ім’я файлу': filename,
                    'Кольорова модель': color_mode,
                    'Формат': img_format,
                    'Кількість каналів': channels,
                    'Розмір у MB': size_mb,
                    'Ширина': width,
                    'Висота': height,
                    'Повний шлях': full_path,
                    'Зменшене зображення': temp_img_path
                }, ignore_index=True)

    # Відображення зменшених зображень з додатковою інформацією
    for index, row in image_data.iterrows():
        display(PILImage.open(row['Зменшене зображення']))
        display(f'Ім’я файлу: {row["Ім’я файлу"]}')
        display(f'Кольорова модель: {row["Кольорова модель"]}')
        display(f'Формат: {row["Формат"]}')
        display(f'Кількість каналів: {row["Кількість каналів"]}')
        display(f'Розмір у MB: {row["Розмір у MB"]}')
        display(f'Ширина: {row["Ширина"]}')
        display(f'Висота: {row["Висота"]}')
        display(f'Повний шлях: {row["Повний шлях"]}')

# Приклад використання функції
folder_path = 'D:\Навчання\ХПІ\Pyton\L4'
process_images(folder_path)
```
